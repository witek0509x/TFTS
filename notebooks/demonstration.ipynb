```python
from generators.nn_generator import NNGenerator
dataloader = NNGenerator(batch_size=1, series_length=40, subseries_length=1, stride=1, store_params=True).get_dataloader()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Create a colormap
num_colors = 50 // 4 + 1  # Number of unique colors needed
colormap = cm.get_cmap("viridis", num_colors)

# Plot with colors based on i // 4
for i in range(20):
    datapoint = next(enumerate(dataloader))
    color = colormap(i // 4 * 5)  # Assign color based on i // 4
    plt.plot(datapoint[1][0].flatten().cpu().numpy(), color=color, alpha=0.3)

# Add legend for the groups
for j in range(num_colors):
    plt.plot([], [], color=colormap(j), label=f"Group {j}")
# plt.legend()
plt.title("Random dynamics paths")
plt.show()
```
